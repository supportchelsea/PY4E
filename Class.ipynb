```
class character():
    def __init__(self, total_move=0, total_damage=0):
        self.total_damage = total_damage
        self.total_move = total_move
    def move(self, step=1):
        self.total_move += step
    def attack(self, damage=1):
        self.total_damage += damage
```

```
class worrior(character):
  def attack(self,damage = 1, add_damage = 2):
    character.attack(self,damage+add_damage)
  def dash(self,dash_dist=3):
    character.move(self,dash_dist)
```

```
worrior_instance = worrior()
worrior_instance.move()
worrior_instance.move()
worrior_instance.attack
```

```
class worrior(character):
    def attack(self, damage=1, add_damage=2):
        character.attack(self, damage+add_damage)
    def dash(self, dash_dist=3): # 돌진 메서드      
        character.move(self, dash_dist)

worrior_instance = worrior()
worrior_instance.move()
worrior_instance.move()
worrior_instance.attack()

print('이동2회, 공격1회')
print('\t총 이동횟수: {}, 총 데미지 {}'.format( worrior_instance.total_move, worrior_instance.total_damage ) )
print(' ')
worrior_instance.dash()
worrior_instance.attack()

print('돌진 1회, 공격 1회')
print('\t총 이동거리: {}, 총 데미지 {}'.format( worrior_instance.total_move, worrior_instance.total_damage ) )
```
