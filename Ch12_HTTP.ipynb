**PART 3. 파이썬을 이용해 웹 데이터 읽어오기
**


```
import socket

mysock = socket.socket(socket.AF_INET,socket.SOCK_STREAM)
# 소켓라이브러리를 가져와서 소켓을 만듬
# 소켓 : 아직 아무데이터하고도 연결되지 않은 데이터 핸들

mysock.connect(('data.pr4e.org',80)) #연결 대상 = (도메인 이름, 포트 번호)-> 튜플

cmd = 'GET http://data.pr4e.org/romeo.txt HTTP/1.0\r\n\r\n'.encode()
#encode를 통해, 유니코드를 utf-8로

mysock.send(cmd)

while True:
  data = mysock.recv(512)
  if (len(data)<1):
    break
  print(data.decode(),end='') #utf-8를 유니코드로
mysock.close()
# 매번 512개씩 받고 반복문 
```

**PART 4. 문자를 표현하는 방법 및 인코딩과 디코딩
**

- ASCII : 아스키 코드는 1 byte로 영문자와 숫자, 그리고 일부 특수문자들을 표현

ord() 함수 = 각각의 문자에 대한 아스키 코드 값 확인 가능


- 유니코드 : 대표적으로 UTF-8

파이썬 2.x 버전에서는 유니코드로 나타내고 싶으면 별도의 문자열 앞에 'u'라는 문자

파이썬 3.x 버전부터는 기본적으로 문자열이 유니코드로 저장





```
print(ord('H'))
print(ord('e'))
print(ord('\n'))
```

```
# Python 3.5.1
x = '이광춘'
print(type(x))
# <class 'str'>
x = u'이광춘'
print(type(x))
# <class 'str'>
```

**PART 5. urllib를 이용해 웹 데이터 읽어오기**

urllib 라이브러리 사용



```
import urllib.request, urllib.parse, urllib.error

fhand = urllib.request.urlopen('http://data.pr4e.org/romeo.txt')
for line in fhand:
    print(line.decode().strip())
```

```
import urllib.request, urllib.parse, urllib.error

fhand = urllib.request.urlopen('http://data.pr4e.org/romeo.txt')


counts = dict()
for line in fhand:
  words = line.decode().split()
  for word in words:
    counts[word] = counts.get(word,0)+1
print(counts)
  
```

**PART 6. BeautifulSoup를 이용한 웹 데이터 스크래핑
**

```
import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup

url = input('Enter -')
html = urllib.request.urlopen(url).read()
soup = BeautifulSoup(html,'html.parser')

# 검색
tags = soup('a')
for tag in tags:
  print(tag.get('href',None))
```

**실습**

```
import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup
import ssl

# SSL 인증서 에러를 무시하기 위해
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE

url = input('Enter - ')
html = urllib.request.urlopen(url,context = ctx).read()
soup = BeautifulSoup(html,'html.parser')

tags = soup('a') #태그가 담긴 리스트를 반환 - 태그 순환
for tag in tags:
  print(tag.get('href',None))

```
